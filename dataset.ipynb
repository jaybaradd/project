```python
import cv2
import os
from mtcnn import MTCNN
from PIL import Image

# Initialize the MTCNN face detector
detector = MTCNN()
yourName = str(input("Name: ")).lower()
# Folder to save detected face images
output_folder = "/Users/rushikesh/Desktop/Jay/vscode_foml/project/facedata/"+ yourName
os.makedirs(output_folder, exist_ok=True)

# Initialize the webcam
target_images = 800
count = 320  # Initialize count

# Initialize the webcam
cap = cv2.VideoCapture(0)

while cap.isOpened() and count < target_images:
    ret, frame = cap.read()
    if not ret:
        print("Failed to grab frame.")
        break

    # Convert frame to RGB (MTCNN expects RGB input)
    rgb_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)

    # Detect faces in the frame
    faces = detector.detect_faces(rgb_frame)

    for face in faces:
        # Get bounding box coordinates
        x, y, width, height = face['box']
        x2, y2 = x + width, y + height

        # Draw bounding box on the original frame
        cv2.rectangle(frame, (x, y), (x2, y2), (0, 255, 0), 2)

        # Extract face from the frame and save it
        face_crop = rgb_frame[y:y2, x:x2]
        if face_crop.size != 0:  # Ensure face crop is not empty
            face_img = Image.fromarray(face_crop)
            face_img.save(os.path.join(output_folder, f"face_{count}.jpg"))
            count += 1  # Increment count after saving

        # Break the loop if the target number of images is reached
        if count >= target_images:
            break

    # Display the frame with bounding boxes
    cv2.imshow('Face Detection', frame)

    # Press 'q' to exit early
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release the capture and close OpenCV windows
cap.release()
cv2.destroyAllWindows()

print(f"Captured {count} face images.")
```

```python
import os
import shutil

# Path to the main folder containing multiple subfolders with images
source_folder = "/Users/rushikesh/Desktop/Jay/vscode_foml/lfw-deepfunneled"
# Path to the target folder where all images will be transferred
target_folder = "/Users/rushikesh/Desktop/Jay/vscode_foml/project/facedata/notjay"

# Create the target folder if it doesn't exist
os.makedirs(target_folder, exist_ok=True)

# Loop through each subfolder in the source folder
for root, dirs, files in os.walk(source_folder):
    for file in files:
        # Check if the file is an image based on its extension
        if file.lower().endswith(('.png', '.jpg', '.jpeg', '.bmp', '.gif')):
            # Construct the full path to the image
            source_file_path = os.path.join(root, file)
            # Construct the path for the image in the target folder
            target_file_path = os.path.join(target_folder, file)

            # Check for filename conflicts and modify the filename if necessary
            counter = 1
            while os.path.exists(target_file_path):
                # If the filename already exists in the target folder, add a suffix
                base, ext = os.path.splitext(file)
                new_filename = f"{base}_{counter}{ext}"
                target_file_path = os.path.join(target_folder, new_filename)
                counter += 1

            # Move the file to the target folder
            shutil.move(source_file_path, target_file_path)

print("All images have been transferred to the target folder.")
```

```python
import os
from mtcnn import MTCNN
from PIL import Image
import numpy as np

# Initialize the MTCNN detector
detector = MTCNN()

# Path to the folder containing images
input_folder = '/Users/rushikesh/Desktop/Jay/vscode_foml/project/facedata/notjay'
output_folder = '/Users/rushikesh/Desktop/Jay/vscode_foml/project/facedata/notjay_cropped'

# Create the output folder if it doesn't exist
os.makedirs(output_folder, exist_ok=True)

# Iterate through each image in the input folder
for filename in os.listdir(input_folder):
    if filename.lower().endswith(('.png', '.jpg', '.jpeg')):
        # Load the image
        image_path = os.path.join(input_folder, filename)
        image = Image.open(image_path)
        image_rgb = image.convert('RGB')  # Convert to RGB if needed
        
        # Detect faces in the image
        faces = detector.detect_faces(np.array(image_rgb))
        
        # Loop through each detected face
        for i, face in enumerate(faces):
            # Get the bounding box coordinates for each face
            x, y, width, height = face['box']
            
            # Crop the face from the image
            face_crop = image_rgb.crop((x, y, x + width, y + height))
            
            # Save the cropped face
            output_filename = f"{os.path.splitext(filename)[0]}_face{i+1}.jpg"
            output_path = os.path.join(output_folder, output_filename)
            face_crop.save(output_path)
            print(f"Saved cropped face to {output_path}")

print("All faces have been cropped and saved.")
```
